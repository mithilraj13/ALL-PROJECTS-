```python
def generate_dummy(size):
    data = []
    for i in range(size):
        temp = []
#         name append
        temp.append(names[random.randint(0, 105)][0])
#         age append
        temp.append(random.randint(20, 50))
#         gender append
        last_char = temp[0][-1]
        if last_char in ['a', 'i']:
            temp.append('F')
#             year append
            temp.append(random.randint(2016, 2019))
            temp.append(random.randint(6000, 16000)/100)
            temp.append(random.randint(1000, 12000)/100)
        else:
            temp.append('M')
#             year append
            temp.append(random.randint(2016, 2019))
#         serum iron append
            temp.append(random.randint(6000, 16000)/100)
            temp.append(random.randint(2000, 25000)/100)
        temp.append(random.randint(6000, 12000)/100)
        temp.append(random.randint(50, 500)/100)
        temp.append(random.choice([1, 0, 0, 0, 0, 0, 0, 0,]))
        temp.append(random.choice([1, 0, 0, 0, 0, 0, 0, 0,]))
        temp.append(0)
#         print(temp)
        data.append(temp)
    return data
data1 = generate_dummy(1000)
```

```python
print(data1[0],
len(data1))
```

```python
def push_into_stuff(new_df, dum_data):
    n = len(new_df.index)
    print(f'Number of elements in df are {n}')
    for i in range(len(dum_data)):
        r = random.randint(0, n+i-1)
        print((r + (r+1))/2)
        new_df.loc[(r + (r+1))/2] = dum_data[i]
        new_df = new_df.sort_index().reset_index(drop=True)
    return new_df
df = push_into_stuff(df, data1)
```
